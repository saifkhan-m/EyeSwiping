```python
import json
import pandas as pd
import numpy as np
```

```python
folderWithTestTrainData='../FinalEvaluation/data/global3/augmented/'
```

```python
def train_test_Data(folder):
    with open(folder+'train.json', "r") as train_file:
        inputdata = json.load(train_file)
    train_data = np.array([[pd.Series(tuple(data['pupilListSmoothed'][:280])),str(data['type'])] for data in inputdata])
    train_data=pd.DataFrame(data=train_data, columns=["att", "label"])
    
    with open(folder+'test.json', "r") as test_file:
        inputdata = json.load(test_file)
    test_data = np.array([[pd.Series(tuple(data['pupilListSmoothed'][:280])),str(data['type'])] for data in inputdata])
    test_data=pd.DataFrame(data=test_data, columns=["att", "label"])
    
    return pd.DataFrame(train_data['att']), train_data['label'], pd.DataFrame(test_data['att']), test_data['label']
```

```python
X_train, y_train, X_test, y_test = train_test_Data(folderWithTestTrainData)
print(type(X_train), type(y_train), X_test.shape, y_test.shape)
```

```python
from sktime.utils.data_container import tabularize
from sklearn.ensemble import RandomForestClassifier

X_train_tab = tabularize(X_train)
X_test_tab = tabularize(X_test)

X_train_tab['label']=y_train
X_test_tab['label']=y_test
X_train_tab.head()
X_test_tab.head()
```

```python
import arff
arff.dump('train.arff'
      , X_train_tab.values
      , relation='eyeswipe'
      , names=X_train_tab.columns)
```
